## 从百度图片下载图片

```python
import urllib
import re

# http://image.baidu.com/channel?c=摄影&t=风景&s=1
url = 'http://image.baidu.com/channel?c=%E6%91%84%E5%BD%B1&t=%E9%A3%8E%E6%99%AF&s=1'
```

```python
f = urllib.urlopen(url)

data = f.read()
```

```python
data
```

### 图片 Json 数据

```python
re_imgs = re.compile(r'.*app.loadImgs\((.*)')
imgs = re.search(re_imgs, data)
```

```python
imgs.group(1)
```

### 直接找图片 URL

```python
re_imgs = re.compile(r'.*?"imageUrl"\s*:\s*"(http://.*?)"')
imgs = re.findall(re_imgs, data)
```

```python
imgs
```

```python
len(imgs)
```

### 下载图片

```python
import os

_filename = lambda s: os.path.join('images', os.path.split(s)[1])

os.mkdir('images')
urllib.urlretrieve(imgs[0], _filename(imgs[0]))
```

```python
def _download(s):
    print('downloading %s' % s)
    urllib.urlretrieve(s, _filename(s))
    return _filename(s)

[_download(i) for i in imgs]
```
